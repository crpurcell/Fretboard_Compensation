## Physics of a Stretched String

Notes plucked on a a ukulele (for instance) cause the string to vibrate in its fundamental mode: nodes at the fixed end-points (saddle, fret or nut) and an anti-node in the middle. The wavelength of the note is fixed and simply given by $\lambda = 2\,L$, where $L$ is the length between the end-points. For an ideal string the frequency (in Hz) of the note is directly related to tension force $\tau$ (in *Newtons*, N) via 

$$f = \frac{1}{2L}\sqrt{\frac{\tau}{\mu}}~~~\ldots~~~(1),$$

where $\mu = \tfrac{m}{l}$ is the mass per unit length of the string (in ${\rm kg\,m^{-1}}$). When a string is stretched beyond its slack length $L_s$ it experiences an internal tension force that opposes the stretching force. Provided the string material is not deformed too much, the magnitude of this force is related to the fundamental elasticity of the string $\gamma$ (in ${\rm N\,m^{-2}\equiv 1\,Pa}$) and its' cross-sectional area $A$ (in ${\rm m^2}$). For a string stretched to a length $L>L_s$, the tension is given by

$$\tau = \gamma\,A\left(\frac{L - L_s}{L_s}\right)~~~\ldots~~~(2).$$

The elasticity $\gamma$ is also known as *Young's Modulus* and encodes how 'stretchy' a material is. For example, steel has $\gamma\approx 200$ GPa (high tension for a small stretch) and nylon has $\gamma\approx 4$ GPa (low tension for a small stretch). Equation (1) can be re-written to solve for tension on a captive string of fixed length $L$ vibrating at fundamental frequency $f$:

$$\tau = \mu\,(2\,L\,f)^2~~~\ldots~~~(3)$$

and Equation (2) to calculate the new length $L$ after applying a tension force $\tau$ to a string held fixed at one end:

$$L = L_s\left(\frac{\tau}{\gamma\,A}+1\right)~~~\ldots~~~(4).$$

### The Effect of Fretting a String

Equations 1 - 4 can be used to relate changes in tension, length and frequency in *different situations*. For a **string of fixed length whose tension is changed** by $\Delta \tau$ (i.e., the situation when an open string on a musical instrument is being tuned) the fractional frequency change $\tfrac{\Delta f}{f}$ is

$$\frac{\Delta f}{f} = \sqrt{1 +\frac{\Delta \tau}{\tau}}-1~~~~~~{\rm and~inverse}~~~~~~\frac{\Delta \tau}{\tau} = \left(1+\frac{\Delta f}{f}\right)^2 -1~~~\ldots~~~(5).$$

For a **string whose length is changed via stretching** by $\Delta L$ (e.g., by moving the end-points further apart, or by bending the string around a third point), the fractional increase in tension is given by 

$$\frac{\Delta \tau}{\tau} = \frac{\Delta L}{L - L_s}~~~\ldots~~~(6).$$

By combining Equations (5) and (6) we get an expression for the fractional change in frequency due to the change in tension because of stretching, *while keeping the vibrating length fixed*:

$$\frac{\Delta f}{f} = \sqrt{1 +\frac{\Delta L}{L - L_s}}-1~~~\ldots~~~(7).$$

Here, $L_s$ can be calculated from Equation (4) and $L$ is the length of the open string (the scale length in the case of a musical instrument). This is the situation when a player depresses a string to a fret: the string is stretched causing the tension to increase, but the vibrating length is fixed by the fret-to-saddle distance. The extra tension causes the note to sound at a higher frequency than ideal. Values for $\Delta L$ and hence $\tfrac{\Delta f}{f}$ vary across the fretboard, causing intonation problems.

## The Effect of String Stiffness

An ideal sting is assumed to bend without resisting the bending force. In the real world the *stiffness* of the string resists the bending force and also means that the deformation of the string starts slightly offset from the bending fulcrum, or end-point. As the stiffness is increased, the result is to raise the pitch of the plucked note compared to an ideal string, with the frequency of higher overtones increasing more (see [*Philip. M. Morse: Vibration and Sound, Chapter IV, Section 16*](https://archive.org/details/VibrationAndSound/page/n187)).

The frequency of a plucked note is given by a modified version of Equation 1, with an extra multiplying term (in square brackets):

$$f = \frac{1}{2L}\sqrt{\frac{\tau}{\mu}}   \left[ 1 + \frac{2}{L}\, \sqrt{\frac{\gamma\,A\,\kappa^2}{T}} + \left(4 + \frac{n^2\,\pi^2}{2}\right)\,\frac{\gamma\,A\,\kappa^2}{T\,L^2} \right]~~~\ldots~~~(8).$$

Here $\mu = \tfrac{m}{l} = \rho\,A$  is the mass per unit length as before, $A$ is the cross-sectional area of the string, $\kappa$ is a quantity called the radius of gyration, which is $\kappa \approx \tfrac{\rm radius}{2}$ for an unwound string. The notation can be simplified by writing

$$\alpha = 4 + \frac{n^2\,\pi^2}{2}~~~~~~{\rm and}~~~~~~\beta = \sqrt{\frac{\gamma\,A\,\kappa^2}{T}}$$

so that Equation 8 becomes

$$f = \frac{1}{2L}\sqrt{\frac{\tau}{\mu}} \left[ 1 + \frac{2\,\beta}{L} + \alpha\,\frac{\beta^2}{L^2} \right]~~~\ldots~~~(9).$$

This formula reduces to Equation 1 as the 'stiffness constant' ($\,\gamma\,A\,\kappa^2$) becomes negligible compared with $T\,L^2$.

## Calculating Fret Positions
Fret positions on a guitar, mandolin or ukulele have traditionally been determined by a geometric progression known as the 'rule of 18', or more precisely: 17.817153. The rule specifies that you repeatedly divide the scale length (distance from the nut to the saddle) by 17.817153 to locate each of the frets with respect to the nut.

This is equivalent to taking a geometric progression from the *saddle* according to:

```
L0 = Scale Length
L1 = L0 / 2**(1/12)
L2 = L1 / 2**(1/12)
L3 = etc ...
```

Using *python*, we can calculate the fret positions like so:

````python
# Necessary imports 
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
````

````python
# Function to calculate fret positions
def calc_fret_positions_vect(scaleLength_m, nutOffset_m=0.0, saddleOffset_m=0.0, 
                             nFrets=18, freqOpen_Hz=440.0, verbose=False):
    """
    Return arrays of vibrating string length and frequency for an instrument
    given a scale length. In the arrays returned, the 1st entry (index = 0) 
    is the nut and the last entry (index = nFrets + 1) is the saddle.
    """
    
    # Calculate the ideal vibrating length from each fret to saddle
    fretIndxArr = np.arange(nFrets + 2)
    vibeLenArr_m = scaleLength_m / (2**(fretIndxArr / 12))
    vibeLenArr_m[0] += nutOffset_m
    vibeLenArr_m += saddleOffset_m
    vibeLenArr_m[-1] = 0.0
    
    # Calculate the fundamental frequency at each fret
    #freqArr_Hz = freqOpen_Hz * vibeLenArr_m[0] / vibeLenArr_m
    freqArr_Hz = np.divide(freqOpen_Hz * vibeLenArr_m[0], 
                           vibeLenArr_m, out=np.zeros_like(vibeLenArr_m), 
                           where=vibeLenArr_m!=0)

    # Calculate the distance from the nut
    distNutArr_m = vibeLenArr_m[0] - vibeLenArr_m
    
    # Print the values if verbose flag set
    if verbose:
        print("Fret  Nut_Distance  Vibrate_Length  Frequency")
        print("              (mm)            (mm)       (Hz)")
        print("-" * 50)
        for i in range(len(distNutArr_m)):
            print(" {:2d}: {:13.1f} {:15.1f} {:10.2f}".format(
                i, distNutArr_m[i]*1000, vibeLenArr_m[i]*1000, freqArr_Hz[i]))
            
    return vibeLenArr_m, freqArr_Hz
````

````python
# Function to calculate the ideal frequency at each fret
def calc_freqs_from_lengths(vibeLenArr_m, freqOpen_Hz=440.0):
    """
    Return an array of frequencies for an ideal string given an array
    of string vibration lengths corresponding to each fret. An ideal
    string is one whose tension does not increase when fretted.
    """
    
    freq_Hz = freqOpen_Hz * vibeLenArr_m[0] / vibeLenArr_m
    
    return freq_Hz
````

````python
# Fundamental parameters of a ukulele.
nFrets = 18
scaleLength_m =  367.0 / 1000
actionNut_m  =     1.0 / 1000
actionFret12_m =   3.5 / 1000
fingerWidth_m =    7.0 / 1000
saddleOffset_m =   0.0 / 1000
nutOffset_m =      0.0 / 1000
````

````python
# Example string tuned to middle A: 440 Hz.
freqOpen_Hz = 440.0

# Calculate the fret positions and frequencies
vibeLenArr_m, freqArr_Hz = calc_fret_positions_vect(
    scaleLength_m=scaleLength_m,
    nutOffset_m=nutOffset_m,
    saddleOffset_m=saddleOffset_m,
    nFrets=nFrets,
    freqOpen_Hz=freqOpen_Hz,
    verbose=True)
````

These fret positions are correct for an *ideal instrument*, that is, one whose strings are not deflected by pressing down on them. In reality, the tension on each string is increased because of the '*clothesline effect*'. This must be calculated for each string seperately, at each fret position, taking into account the string action and width of the finger pressing down. A simple geometric model for a fretted string is shown in the diagram below:

Here the depressing finger is modelled with a flat profile, which is accurate enough for calculations here, as other sources of error likely dominate.

The python code below calculates the *increase in length* due to the clothesline effect. Note that the corresponding increase in tension will depend on string properties (mass-per-unit-length, cross-sectional area) and open tuned frequency.

````python
# Function to calculate the string length when depressed
def calculate_fretted_length(vibeLenArr, actionNut=1.5/1000, actionFret12=3.5/1000,
                             fingerWidth=7.0/1000, verbose=False):
    """
    Return an array of deflected lengths corresponding to string being depressed behind each fret.
    """    
    
    # Calculate the slope of the open string
    distNutArr = vibeLenArr[0] - vibeLenArr
    print(distNutArr)
    x1 = 0
    x2 = distNutArr[12]
    y1 = actionNut
    y2 = actionFret12
    m = (y2 - y1) / (x2 - x1)
    
    # Calculate the action at the saddle
    scaleLength = distNutArr[12] * 2
    actionSaddle = m * scaleLength + actionNut    
    
    # Calculate Lprime: the length of the string projected to the fingerboard
    Lprime = np.sqrt(scaleLength**2.0 - (actionSaddle - actionNut)**2.0)
        
    # Calculate the distance to the top of each finger press.
    # Clip finger width if the distance is less than the distance to the previous fret.
    distPressArr = distNutArr - fingerWidth
    distPressArr[1:] = np.where(distPressArr[1:] < distNutArr[:-1], 
                                distNutArr[:-1], distPressArr[1:])
    distPressArr[0] = 0.0
    fingerWidthArr = distNutArr - distPressArr
    
    # Calculate L1, the string length above the finger press
    L1 = np.sqrt(actionNut**2.0 + distPressArr**2.0)
    
    # Calculate L2, the string length below the finger press
    L2 = np.sqrt(actionSaddle**2.0 + (Lprime - distNutArr)**2.0)
    
    # Deflected length: nut-to-finger + fingerWidth + fret-to-saddle
    stringDeflArr = L1 + fingerWidthArr + L2
    stringDeflArr[0] = 0.0
    
    if verbose:
        print("Saddle Action    = {:6.2f}".format(actionSaddle))
        print("Projected Length = {:6.2f}".format(Lprime))
    
    return stringDeflArr
````

````python
# Calculate the deflection for a thin and normal finger.
# Assume default fretboard parameters.
stringDeflThinArr_m = calculate_fretted_length(vibeLenArr_m, actionNut_m, actionFret12_m, 3.0/1000)
stringDeflArr_m = calculate_fretted_length(vibeLenArr_m, actionNut_m, actionFret12_m, fingerWidth_m)

# Calculate the increase in length, delta_L
deltaThinL_m = stringDeflThinArr_m - vibeLenArr_m
deltaL_m = stringDeflArr_m - vibeLenArr_m
````

````python
# Plot the deflection for a thin and normal finger
fretNumArr = np.arange(1, len(deltaL_m[1:])+1)
plt.plot(fretNumArr, deltaThinL_m[1:], marker="o", label="Thin")
plt.plot(fretNumArr, deltaL_m[1:], marker="^", label="Normal")
leg = plt.legend()
plt.xlabel("Fret Number")
plt.ylabel("$\Delta L (m)$")
````

Now we need to calculate the fractional change in frequency due to the fretted tension for each position. This can be found directly from Equation (7), provided ($\gamma\,A$) is known for the string. This fractional change is applied to the fundamental frequency of the fretted note.

````python
# Function to calculate the fractional change in frequency considering length only
def calc_delta_freq(freqOpen_Hz, lengthOpen_mm, massPerLength_kgm, stringDiameter_mm, deltaL_mm):
    """
    Return an array of fractional frequency change given the open string frequency, 
    scale-length, string properties and the increase in string length due to bending.
    """
    
    # Convert to SI
    lengthOpen_m = lengthOpen_mm / 1000
    deltaL_m = deltaL_mm / 1000
    stringDiameter_m = stringDiameter_mm / 1000
    
    # Calculate the tension in the open string
    tau_N = massPerLength_kgm * np.power(2 * lengthOpen_m * freqOpen_Hz, 2)

    # Calculate the slack length
    area_m2 = np.pi * (stringDiameter_m / 2)**2.0
    lSlack_m = lengthOpen_m / (tau_N / (elasticity_Pa * area_m2) + 1)

    # Calculate fractional change in frequency
    freqFrac = np.sqrt(1 + deltaL_m[1:]/(lengthOpen_m - lSlack_m)) - 1

    return freqFrac
````

````python
# Function to calculate the fractional change in frequency considering length & string stiffness
def calc_delta_freq_full(freqOpen_Hz, lengthOpen_mm, massPerLength_kgm, stringDiameter_mm, deltaL_mm, elasticity_Pa):
    """
    Return an array of fractional frequency change given the open string frequency, 
    scale-length, string properties and the increase in string length due to bending.
    """
    
    # Convert to SI
    lengthOpen_m = lengthOpen_mm / 1000
    deltaL_m = deltaL_mm / 1000
    stringDiameter_m = stringDiameter_mm / 1000
    stringRadius_m = stringDiameter_m / 2
    
    # Calculate the tension in the open string
    tau_N = massPerLength_kgm * np.power(2 * lengthOpen_m * freqOpen_Hz, 2)

    # Calculate the slack length
    area_m2 = np.pi * stringRadius_m**2.0
    lSlack_m = lengthOpen_m / (tau_N / (elasticity_Pa * area_m2) + 1)

    # Calculate fractional change in frequency because of length changes
    freqFracLen = np.sqrt(1 + deltaL_m[1:]/(lengthOpen_m - lSlack_m)) - 1

    # Calculate the alpha and beta stretch terms
    n = 1.0
    alpha = 4 + (n**2 * np.pi**2) / 2
    kappa = stringRadius_m / 2
    beta = np.sqrt(elasticity_Pa * area_m2 * kappa**2 / tau_N)

    # Calculate fractional change in frequency because of stiffness
    freqMultStiff = 1 + 2 * beta / lengthOpen_m + alpha * beta**2 / lengthOpen_m**2
    freqFracStiff = ((freqMultStiff * freqOpen_Hz) - freqOpen_Hz) / freqOpen_Hz
    
    return freqFracLen, freqFracStiff
````

````python
# String properties
massPerLength_kgm = 1e-4
stringDiameter_mm = 0.5
elasticity_Pa = 4e9

# Calculate the frequency increase
freqFrac = calc_delta_freq(freqOpen_Hz, scaleLength_mm, massPerLength_kgm, stringDiameter_mm, deltaL_mm)
freqFracLen, freqFracStiff = calc_delta_freq_full(freqOpen_Hz, scaleLength_mm, massPerLength_kgm, stringDiameter_mm, deltaL_mm, elasticity_Pa)

# Calculate the absolute change in frequency
deltaF1_Hz = freqFracLen * freqArr_Hz[1:]
deltaF2_Hz = freqFracStiff * freqArr_Hz[1:]
freqFrac = freqFracLen + freqFracStiff
deltaF_Hz = freqFrac * freqArr_Hz[1:]
````

````python
# Plot the absolute change in frequency
plt.plot(fretNumArr, deltaF1_Hz, marker="o", label="$\Delta$ Frequency Length")
plt.plot(fretNumArr, deltaF2_Hz, marker="o", label="$\Delta$ Frequency Stiff")
plt.plot(fretNumArr, deltaF_Hz, marker="o", label="$\Delta$ Frequency Total")
leg = plt.legend()
plt.xlabel("Fret Number")
plt.ylabel("$\Delta f (Hz)$")
````

````python
# Plot the fractional frequency increase
plt.plot(fretNumArr, freqFrac*1000, marker="o", label="$\Delta$ f / f")
leg = plt.legend()
plt.xlabel("Fret Number")
plt.ylabel("Fraction x1000")
````

````python
# Calculate the new frequency array
freqNew_Hz = freqArr_Hz.copy()
freqNew_Hz[1:] += deltaF_Hz

# Calculate the change in cents
centArr = 1200 * np.log2(freqNew_Hz / freqArr_Hz)

# Plot the frequency increase in cents
plt.plot(fretNumArr, centArr[1:], marker="o", label="$\Delta$ Frequency")
leg = plt.legend()
plt.xlabel("Fret Number")
plt.ylabel("$\Delta f (cents)$")
````

## Correcting for the Frequency Shift
We can take two approaches at this point. 1) Calculate custom fret positions and length scales for each string to compensate for the frequency shifts, or 2) retain the 'rule of 18' fret positions, but shift the upper and lower end-points (nut and saddle, respectively). The second approach is easier, can be applied to existing instruments and can correct intonation well enough for human perception. This is simply an optimisation problem akin to fitting a curve to data - $\chi^2$ minimisation.

We need to define an error function that takes as arguments the fixed and free parameters of the instrument, and returns the $\chi^2$ (a measure of the 'distance' between the current model and the ideal). The parameters will be as follows:

| Parameter | Fit | Description |
| --- | --- | --- |
| scaleLength_mm | [FIXED] | scale-length of the instrument |
| nFrets | [FIXED] | number of frets |
| actionNut_mm | [FIXED] | height of the strings at the nut |
| actionFret12_mm | [FIXED] | height of the strings at fret 12 |
| fingerWidth_mm | [FIXED] | width of the depressing finger |
| freqOpen_Hz | [FIXED] | frequency of the open string |
| massPerLength_kgm | [FIXED] | mass-per-unit-length of the string |
| stringDiameter_mm | [FIXED] | diameter of the string |
| elasticity_Pa | [FIXED] | elasticity of the string |
| nutOffset_mm | **[FREE]** | offset of the nut position |
| saddleOffset_mm | **[FREE]** | offset of the saddle position |

Because human perception of musical tone is logarithmic, we should optimise $\Delta f_{\rm cents}$, rather than $\Delta f = (f_{\rm new} - f_{\rm old})$. That is, we minimise the value of

$$\chi^2 = \sum_{0}^{N_{\rm frets}}\,(\Delta f_{\rm cents})^2.$$

by adjusting saddle and nut offset.

````python
# Function to 'model' frequency offset of a fretted string
def model(scaleLength_mm, nFrets, actionNut_mm, actionFret12_mm, fingerWidth_mm, freqOpen_Hz, 
          massPerLength_kgm, stringDiameter_mm, elasticity_Pa, nutOffset_mm, saddleOffset_mm):
    
    # Ideal instrument, without offsets
    distNutIdealArr_mm, vibeLenIdealArr_mm, freqIdealArr_Hz = \
        calc_fret_positions(scaleLength_mm, nFrets, freqOpen_Hz)
    
    # Instrument with nut offset
    distNutOffArr_mm = distNutIdealArr_mm.copy()
    distNutOffArr_mm[1:] += nutOffset_mm     # Nut offset increases distance nut->frets
    vibeLenOffArr_mm = vibeLenIdealArr_mm.copy()
    vibeLenOffArr_mm[0] += nutOffset_mm      # Nut offset increases length of open string only
    vibeLenOffArr_mm += saddleOffset_mm      # Saddle offset increases length of all positions
    openLength_mm = vibeLenOffArr_mm[0]
    
    # Calculate the deflected length
    stringDeflArr_mm = calculate_fretted_length(distNutOffArr_mm, actionNut_mm, actionFret12_mm,
                                                fingerWidth_mm, saddleOffset_mm)
    
    # Fractional frequency change
    #freqFrac = calc_delta_freq(freqOpen_Hz, openLength_mm, massPerLength_kgm, stringDiameter_mm, deltaL_mm)
    freqFracLen, freqFracStiff = calc_delta_freq_full(freqOpen_Hz, scaleLength_mm, massPerLength_kgm, stringDiameter_mm, deltaL_mm, elasticity_Pa)
    freqFrac = freqFracLen + freqFracStiff
    
    # Calculate the absolute change in frequency
    freqArr_Hz = calc_freqs_from_lengths(vibeLenOffArr_mm, freqOpen_Hz)
    deltaF_Hz = freqFrac * freqArr_Hz[1:]
    
    # Calculate the new frequency array
    freqNew_Hz = freqArr_Hz.copy()
    freqNew_Hz[1:] += deltaF_Hz

    # Calculate the change in cents
    centArr = 1200 * np.log2(freqNew_Hz / freqIdealArr_Hz)
    
    return centArr
````

````python
# Call the model to check it works
centArr = model(scaleLength_mm, nFrets, actionNut_mm, actionFret12_mm, fingerWidth_mm, freqOpen_Hz,
                massPerLength_kgm, stringDiameter_mm, elasticity_Pa, 
                nutOffset_mm=0.0, saddleOffset_mm=0.0)

centSadArr = model(scaleLength_mm, nFrets, actionNut_mm, actionFret12_mm, fingerWidth_mm, freqOpen_Hz,
                massPerLength_kgm, stringDiameter_mm, elasticity_Pa, 
                nutOffset_mm=0.0, saddleOffset_mm=5.0)

centBothArr = model(scaleLength_mm, nFrets, actionNut_mm, actionFret12_mm, fingerWidth_mm, freqOpen_Hz,
                massPerLength_kgm, stringDiameter_mm, elasticity_Pa, 
                nutOffset_mm=-3.5, saddleOffset_mm=1.3)
````

````python
# Plot the frequency increase in cents
plt.plot(fretNumArr, centArr[1:], marker="o", label="Uncorrected")
plt.plot(fretNumArr, centSadArr[1:], marker="o", label="Saddle Corrected")
plt.plot(fretNumArr, centBothArr[1:], marker="o", label="Both Corrected")
plt.axhline(color="grey")
leg = plt.legend()
plt.xlabel("Fret Number")
plt.ylabel("$\Delta f (cents)$")
````
