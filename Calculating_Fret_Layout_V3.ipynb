## Physics of a Stretched String

Notes plucked on a a ukulele (for instance) cause the string to vibrate in its fundamental mode: nodes at the fixed end-points (saddle, fret or nut) and an anti-node in the middle. The wavelength of the note is fixed and simply given by $\lambda = 2\,L$, where $L$ is the length between the end-points. For an ideal string the frequency (in Hz) of the note is directly related to tension force $\tau$ (in *Newtons*, N) via 

$$f = \frac{1}{2L}\sqrt{\frac{\tau}{\mu}}~~~\ldots~~~(1),$$

where $\mu = \tfrac{m}{l}$ is the mass per unit length of the string (in ${\rm kg\,m^{-1}}$). When a string is stretched beyond its slack length $L_s$ it experiences an internal tension force that opposes the stretching force. Provided the string material is not deformed too much, the magnitude of this force is related to the fundamental elasticity of the string $\gamma$ (in ${\rm N\,m^{-2}\equiv 1\,Pa}$) and its' cross-sectional area $A$ (in ${\rm m^2}$). For a string stretched to a length $L>L_s$, the tension is given by

$$\tau = \gamma\,A\left(\frac{L - L_s}{L_s}\right)~~~\ldots~~~(2).$$

The elasticity $\gamma$ is also known as *Young's Modulus* and encodes how 'stretchy' a material is. For example, steel has $\gamma\approx 200$ GPa (high tension for a small stretch) and nylon has $\gamma\approx 4$ GPa (low tension for a small stretch). Equation (1) can be re-written to solve for tension on a captive string of fixed length $L$ vibrating at fundamental frequency $f$:

$$\tau = \mu\,(2\,L\,f)^2~~~\ldots~~~(3)$$

and Equation (2) to calculate the new length $L$ after applying a tension force $\tau$ to a string held fixed at one end:

$$L = L_s\left(\frac{\tau}{\gamma\,A}+1\right)~~~\ldots~~~(4).$$

### The Effect of Fretting a String

Equations 1 - 4 can be used to relate changes in tension, length and frequency in *different situations*. For a **string of fixed length whose tension is changed** by $\Delta \tau$ (i.e., the situation when an open string on a musical instrument is being tuned) the fractional frequency change $\tfrac{\Delta f}{f}$ is

$$\frac{\Delta f}{f} = \sqrt{1 +\frac{\Delta \tau}{\tau}}-1~~~~~~{\rm and~inverse}~~~~~~\frac{\Delta \tau}{\tau} = \left(1+\frac{\Delta f}{f}\right)^2 -1~~~\ldots~~~(5).$$

For a **string whose length is changed via stretching** by $\Delta L$ (e.g., by moving the end-points further apart, or by bending the string around a third point), the fractional increase in tension is given by 

$$\frac{\Delta \tau}{\tau} = \frac{\Delta L}{L - L_s}~~~\ldots~~~(6).$$

By combining Equations (5) and (6) we get an expression for the fractional change in frequency due to the change in tension because of stretching, *while keeping the vibrating length fixed*:

$$\frac{\Delta f}{f} = \sqrt{1 +\frac{\Delta L}{L - L_s}}-1~~~\ldots~~~(7).$$

Here, $L_s$ can be calculated from Equation (4) and $L$ is the length of the open string (the scale length in the case of a musical instrument). This is the situation when a player depresses a string to a fret: the string is stretched causing the tension to increase, but the vibrating length is fixed by the fret-to-saddle distance. The extra tension causes the note to sound at a higher frequency than ideal. Values for $\Delta L$ and hence $\tfrac{\Delta f}{f}$ vary across the fretboard, causing intonation problems.

## The Effect of String Stiffness

An ideal sting is assumed to bend without resisting the bending force. In the real world the *stiffness* of the string resists the bending force and also means that the deformation of the string starts slightly offset from the bending fulcrum, or end-point. As the stiffness is increased, the result is to raise the pitch of the plucked note compared to an ideal string, with the frequency of higher harmonics increasing more (see [*Philip. M. Morse: Vibration and Sound, Chapter IV, Section 16*](https://archive.org/details/VibrationAndSound/page/n187)).

The frequency of a plucked note is given by a modified version of Equation 1, with an extra multiplying term (in square brackets):

$$f = \frac{1}{2L}\sqrt{\frac{\tau}{\mu}}   \left[ 1 + \frac{2}{L}\, \sqrt{\frac{\gamma\,A\,\kappa^2}{\tau}} + \left(4 + \frac{n^2\,\pi^2}{2}\right)\,\frac{\gamma\,A\,\kappa^2}{\tau\,L^2} \right]~~~\ldots~~~(8).$$

Here $\mu = \tfrac{m}{l} = \rho\,A$  is the mass per unit length as before, $A$ is the cross-sectional area of the string, $\kappa$ is a quantity called the radius of gyration, which is $\kappa \approx \tfrac{\rm radius}{2}$ for an unwound string. The notation can be simplified by writing

$$\alpha = 4 + \frac{n^2\,\pi^2}{2}~~~~~~{\rm and}~~~~~~\beta = \sqrt{\frac{\gamma\,A\,\kappa^2}{\tau}}$$

so that Equation 8 becomes

$$f = \frac{1}{2L}\sqrt{\frac{\tau}{\mu}} \left[ 1 + \frac{2\,\beta}{L} + \alpha\,\frac{\beta^2}{L^2} \right]~~~\ldots~~~(9).$$

This formula reduces to Equation 1 as the 'stiffness constant' ($\,\gamma\,A\,\kappa^2$) becomes negligible compared with $\tau\,L^2$.

## Calculating Fret Positions
Fret positions on a guitar, mandolin or ukulele have traditionally been determined by a geometric progression known as the 'rule of 18', or more precisely: 17.817153. The rule specifies that you repeatedly divide the scale length (distance from the nut to the saddle) by 17.817153 to locate each of the frets with respect to the nut.

This is equivalent to taking a geometric progression from the *saddle* according to:

```
L0 = Scale Length
L1 = L0 / 2**(1/12)
L2 = L1 / 2**(1/12)
L3 = etc ...
```

Using *python*, we can calculate the fret positions like so:

````python
# Necessary imports 
from matplotlib import pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import scipy.optimize as op
%matplotlib inline
````

````python
# Function to calculate fret positions and frequencies
def calc_vibelen_freq(scaleLength_m, nFrets=18, freqOpen_Hz=440.0, 
                      nutOffset_m=0.0, saddleOffset_m=0.0, verbose=False):
    """
    Return arrays of vibrating string length and frequency for an ideal
    instrument given a scale length. In the arrays returned, the 1st
    entry [0] is the nut and the last entry [nFrets + 1] is the saddle.
    """
    
    # Calculate the ideal vibrating length from each fret to saddle
    fretIndxArr = np.arange(nFrets + 2)
    vibeLenArr_m = scaleLength_m / (2**(fretIndxArr / 12))
    vibeLenArr_m[0] += nutOffset_m
    vibeLenArr_m += saddleOffset_m
    vibeLenArr_m[-1] = 0.0
    
    # Calculate the fundamental frequency at each fret
    
    with np.errstate(divide='ignore', invalid='ignore'):
        freqArr_Hz = freqOpen_Hz * vibeLenArr_m[0] / vibeLenArr_m
    
    # Calculate the distance from the nut
    distNutArr_m = vibeLenArr_m[0] - vibeLenArr_m
    
    # Print the values if verbose flag set
    if verbose:
        print("Fret  Nut_Distance  Vibrate_Length  Frequency")
        print("              (mm)            (mm)       (Hz)")
        print("-" * 50)
        for i in range(len(distNutArr_m)):
            print(" {:2d}: {:13.1f} {:15.1f} {:10.2f}".format(
                i, distNutArr_m[i]*1000, vibeLenArr_m[i]*1000, freqArr_Hz[i]))
            
    return vibeLenArr_m, freqArr_Hz
````

````python
# Geometric parameters of a ukulele
nFrets =               18
scaleLength_m =     367.0 / 1000
actionNut_m  =        1.5 / 1000
actionFret12_m =      3.5 / 1000
fingerWidth_m =       5.0 / 1000

# String specific parameters
freqOpen_Hz =       440.0
massPerLength_kgm =  1e-4
stringDiameter_m =    0.5 / 1000
elasticity_Pa =       4e9
nutOffset_m =        -2.0 / 1000
saddleOffset_m =      3.0 / 1000
````

````python
# Calculate the ideal fret positions and frequency array
vibeLenIdealArr_m, freqIdealArr_Hz = calc_vibelen_freq(
    scaleLength_m=scaleLength_m,
    nFrets=nFrets,
    freqOpen_Hz=freqOpen_Hz,
    verbose=False)

# Calculate the positions & frequencies including offsets
vibeLenArr_m, freqArr_Hz = calc_vibelen_freq(
    scaleLength_m=scaleLength_m,
    nFrets=nFrets,
    freqOpen_Hz=freqOpen_Hz,
    nutOffset_m=nutOffset_m,
    saddleOffset_m=saddleOffset_m,
    verbose=True)
````

These fret positions are correct for an *ideal instrument*, that is, one whose strings are not deflected by pressing down on them. In reality, the tension on each string is increased because of the '*clothesline effect*'. This must be calculated for each string seperately, at each fret position, taking into account the string action and width of the finger pressing down. Here the depressing finger is modelled with a flat profile, which is accurate enough for calculations here, as other sources of error likely dominate.

The python code below calculates the *increase in length* due to the clothesline effect. Note that the corresponding increase in tension will depend on string properties (mass-per-unit-length, cross-sectional area) and open tuned frequency.

````python
# Function to calculate the string length when depressed
def calculate_fretted_length(vibeLenArr, actionNut=1.5/1000, actionFret12=3.5/1000,
                             fingerWidth=7.0/1000, verbose=False):
    """
    Return an array of total (i.e., open) string lengths 
    corresponding to the string being depressed behind each fret.
    """    
    
    # Calculate the slope of the open string and action at saddle
    x1 = vibeLenArr[12]
    x2 = vibeLenArr[0]
    y1 = actionFret12
    y2 = actionNut
    m = (y2 - y1) / (x2 - x1)
    actionSaddle = actionFret12 - m * vibeLenArr[12]
    
    # Calculate Lprime: the length of the string projected to the fingerboard
    Lprime = np.sqrt(vibeLenArr[0]**2.0 - (actionSaddle - actionNut)**2.0)
    
    # Calculate the clipped finger width if the finger overlaps previous fret
    fretWidthArr = np.zeros_like(vibeLenArr)
    fretWidthArr[1:] = -1 * np.diff(vibeLenArr)
    fingerWidthArr = np.where(fretWidthArr <= fingerWidth, fretWidthArr, fingerWidth)
    fingerWidthArr[-1] = 0
    
    # Calculate L1, the string length above the finger press
    L1 = np.sqrt(actionNut**2.0 + (Lprime - fingerWidthArr - vibeLenArr)**2.0)
    
    # Calculate L2, the string length below the finger press
    L2 = np.sqrt(actionSaddle**2.0 + vibeLenArr**2.0)
    
    # Deflected length: nut-to-finger + fingerWidth + fret-to-saddle
    # Set the nut to the open length and saddle to np.nan
    newLArr = L1 + fingerWidthArr + L2
    newLArr[0] = vibeLenArr[0]
    newLArr[-1] = np.nan
    
    # Some feedback
    if verbose:
        print("Saddle Action    = {:6.2f}".format(actionSaddle))
        print("Projected Length = {:6.2f}".format(Lprime))
    
    return newLArr
````

````python
# Calculate the deflected length for a thin and normal finger.
fingerWidthThin_m = 0.0/1000
newLThinArr_m = calculate_fretted_length(vibeLenArr_m, actionNut_m, 
                                         actionFret12_m, fingerWidthThin_m)
newLArr_m = calculate_fretted_length(vibeLenArr_m, actionNut_m, 
                                     actionFret12_m, fingerWidth_m)
````

````python
# Calculate the increase in length, delta_L
deltaThinL_m = newLThinArr_m - vibeLenArr_m[0]
deltaL_m = newLArr_m - vibeLenArr_m[0]

# Plot the length increase for a thin and normal finger
fretNumArr = np.arange(len(vibeLenArr_m))
ax = plt.subplot(1, 1, 1)
ax.plot(fretNumArr[1:], deltaThinL_m[1:]*1000, marker="o", label="Thin")
ax.plot(fretNumArr[1:], deltaL_m[1:]*1000, marker="o", label="Normal")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
leg = ax.legend()
ax.set_xlabel("Fret Number")
ax.set_ylabel("$\Delta L (mm)$")
````

From the increase in length we can calculate the increase in tension as a function of fret position. This can be found directly from Equation (3), provided ($\gamma\,A$) is known for the string.

````python
# Function to calculate the tension multiplier after fretting the string
def calc_tension_mult(freqOpen_Hz, newLArr_m, massPerLength_kgm, 
                      stringDiameter_m, elasticity_Pa):
    """
    Return an array of (tau + delta_tau) / given the open string frequency, 
    scale-length, string properties and the increase in string length due to bending.
    """
    
    # Calculate the tension in the open string
    lengthOpen_m = newLArr_m[0]
    tauOpen_N = massPerLength_kgm * np.power(2 * lengthOpen_m * freqOpen_Hz, 2)

    # Calculate the slack length
    area_m2 = np.pi * (stringDiameter_m / 2)**2.0
    slackL_m = lengthOpen_m / (tauOpen_N / (elasticity_Pa * area_m2) + 1)
    
    # Calculate the tension in the deflected string
    tauArr_N =  elasticity_Pa * area_m2 * (newLArr_m - slackL_m) / slackL_m
    
    # Frequency multiplier proportional to sqrt(tension)
    return np.sqrt(tauArr_N / tauOpen_N)
````

````python
# Calculate tension multiplier
tauFreqMultArr = calc_tension_mult(freqOpen_Hz, newLArr_m, massPerLength_kgm, 
                                   stringDiameter_m, elasticity_Pa)
````

````python
# Plot the tension multiplier
ax = plt.subplot(1, 1, 1)
ax.plot(fretNumArr[1:], tauFreqMultArr[1:], marker="^", label="Tension Multiplier")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
leg = ax.legend()
ax.set_xlabel("Fret Number")
ax.set_ylabel("Multiplier")
````

Seperately, we can estimate the fractional increase in frequency due to the stiffness of the string

````python
# Calculate the string stiffness multiplier 
def calc_stiff_mult(freqOpen_Hz, vibeLenArr_m, massPerLength_kgm, stringDiameter_m, elasticity_Pa, n=1):
    """
    Calculate the stiffness multiplier: the increase (f + delta_f) / f
    of the frequency of the nth harmonic due to the resistance of the
    string to bending motions. Assume n=1 dominates perception of tuning.
    """
    
    # Calculate the tension in the open string
    lengthOpen_m = vibeLenArr_m[0]
    tau_N = massPerLength_kgm * np.power(2 * lengthOpen_m * freqOpen_Hz, 2)

    # Calculate the slack length
    stringRadius_m = stringDiameter_m / 2
    area_m2 = np.pi * stringRadius_m**2
    lSlack_m = lengthOpen_m / (tau_N / (elasticity_Pa * area_m2) + 1)

    # Calculate the alpha and beta stretch terms
    alpha = 4 + (n**2 * np.pi**2) / 2
    kappa = stringRadius_m / 2
    beta = np.sqrt(elasticity_Pa * area_m2 * kappa**2 / tau_N)

    # Calculate stiffness multiplier
    with np.errstate(divide='ignore', invalid='ignore'):
        stiffFreqMultArr = 1 + 2 * beta / vibeLenArr_m + alpha * beta**2 / vibeLenArr_m**2
        
    # Assuming tuned to open string, so normalise to 1 at nut
    stiffFreqMultArr /= stiffFreqMultArr[0]
    stiffFreqMultArr[stiffFreqMultArr == np.inf] = np.nan
    
    return stiffFreqMultArr
````

````python
# Calculate the stiffness multiplier
stiffFreqMultArr = calc_stiff_mult(freqOpen_Hz, vibeLenArr_m, massPerLength_kgm, stringDiameter_m, elasticity_Pa)

# Calculate the combined multiplier
freqMultArr = stiffFreqMultArr * tauFreqMultArr

# Print out values
print("Fret  Stiffness    Tension      Total")
print("-" * 50)
for i in range(len(freqMultArr)):
    print(" {:2d}: {:10.4f} {:10.4f} {:10.4f}".format(
        i, stiffFreqMultArr[i], tauFreqMultArr[i], freqMultArr[i]))
            
````

Now we have calculated arrays of $\frac{f + \Delta f}{f}$ for both effects (increased tension and string stiffness). We can plot their values as a function of fret number:

````python
# Plot both multipliers
ax = plt.subplot(1, 1, 1)
ax.plot(fretNumArr[1:], tauFreqMultArr[1:], marker="o", label="Tension")
ax.plot(fretNumArr[1:], stiffFreqMultArr[1:], marker="^", label="Stiffness")
ax.plot(fretNumArr[1:], freqMultArr[1:], marker="<", label="Combined")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
leg = ax.legend()
ax.set_xlabel("Fret Number")
ax.set_ylabel("Multiplier")
````

The response of the human ear is logarithmic and musicians measures differences in frequency in 'cents', which are defined as:

$$\Delta C = 1200 \times {\rm log}_2\left(\frac{f_2}{f_1}\right)$$

This leads to a plot of the intonation error:

````python
# Calculate the deviation from the ideal frequencies
freqNewStiffArr_Hz = freqArr_Hz * stiffFreqMultArr
freqNewTauArr_Hz = freqArr_Hz * tauFreqMultArr
freqNewArr_Hz = freqArr_Hz * freqMultArr

# Calculate the difference in cents
stiffCentArr = 1200 * np.log2(freqNewStiffArr_Hz / freqIdealArr_Hz)
tauCentArr = 1200 * np.log2(freqNewTauArr_Hz / freqIdealArr_Hz)
centArr = 1200 * np.log2(freqNewArr_Hz / freqIdealArr_Hz)

# Plot the frequency increase in cents
ax = plt.subplot(1, 1, 1)
ax.plot(fretNumArr[1:], tauCentArr[1:], marker="o", label="Tension")
ax.plot(fretNumArr[1:], stiffCentArr[1:], marker="^", label="Stiffness")
ax.plot(fretNumArr[1:], centArr[1:], marker="<", label="Combined")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.axhline(color="grey")
leg = ax.legend()
ax.set_xlabel("Fret Number")
ax.set_ylabel("$\Delta f (cents)$")
````

## Correcting for the Frequency Shift
We can take two approaches at this point. 1) Calculate custom fret positions and length scales for each string to compensate for the frequency shifts, or 2) retain the 'rule of 18' fret positions, but shift the upper and lower end-points (nut and saddle, respectively). The second approach is easier, can be applied to existing instruments and can correct intonation well enough for human perception. This is simply an optimisation problem akin to fitting a curve to data - $\chi^2$ minimisation.

We need to define an error function that takes as arguments the fixed and free parameters of the instrument and string, and returns the $\chi^2$ (a measure of the 'distance' between the current model and the ideal). The parameters will be as follows:

| Parameter | Fit | Description |
| --- | --- | --- |
| scaleLength_m | [FIXED] | ideal scale-length of the instrument |
| nFrets | [FIXED] | number of frets |
| actionNut_m | [FIXED] | height of the strings at the nut |
| actionFret12_m | [FIXED] | height of the strings at fret 12 |
| fingerWidth_m | [FIXED] | width of the depressing finger |
| freqOpen_Hz | [FIXED] | frequency of the open string |
| massPerLength_kgm | [FIXED] | mass-per-unit-length of the string |
| stringDiameter_m | [FIXED] | diameter of the string |
| elasticity_Pa | [FIXED] | elasticity of the string |
| nutOffset_m | **[FREE]** | offset of the nut position towards headstock |
| saddleOffset_m | **[FREE]** | offset of the saddle position away from headstock  |

Because human perception of musical tone is logarithmic, we should optimise $\Delta C$, rather than $\Delta f = (f_{\rm new} - f_{\rm old})$. That is, we adjust the saddle and nut offset to minimise the value of

$$\chi^2 = \sum_{i\,=\,0}^{N_{\rm frets}}\,\left(\frac{\Delta C_i}{W_i} \right)^2.$$

Here $W$ is a vector of weights that can be unity, or set to some other function in order to prioritise corrections for different parts of the fretboard. First we define a model function that will return an array of intonation error as a function of fret number. To work with the optimiser, this model function should return *another function* that takes the free parameters as a vector *p*.

````python
# Define a function to calculate the frequency offset of a fretted string
def get_model_func(scaleLength_m, nFrets, actionNut_m, actionFret12_m,
                   fingerWidth_m, freqOpen_Hz, massPerLength_kgm, 
                   stringDiameter_m, elasticity_Pa):
    """
    Get a model function with fixed instrument & string parameters
    """
    
    # Return function takes only a vector of free parameters
    def calc_deltafreq_cent(p):
        """
        Model function to calculate intonation given nut & bridge offset.
        
        p = [nutOffset_m, saddleOffset_m]
        
        """
    
        # Assign vector to named variables
        nutOffset_m, saddleOffset_m = p
        
        # Calculate the ideal fret positions and frequency array
        vibeLenIdealArr_m, freqIdealArr_Hz = \
            calc_vibelen_freq(scaleLength_m, nFrets, freqOpen_Hz)

        # Calculate the positions & ideal frequencies including offsets
        vibeLenArr_m, freqArr_Hz = \
            calc_vibelen_freq(scaleLength_m, nFrets, freqOpen_Hz, 
                              nutOffset_m, saddleOffset_m)
    
        # Calculate the deflected length
        newLArr_m = calculate_fretted_length(vibeLenArr_m, actionNut_m, 
                                             actionFret12_m, fingerWidth_m)
    
        # Calculate tension multiplier
        tauFreqMultArr = calc_tension_mult(freqOpen_Hz, newLArr_m, massPerLength_kgm, 
                                           stringDiameter_m, elasticity_Pa)
    
        # Calculate the stiffness multiplier
        stiffFreqMultArr = calc_stiff_mult(freqOpen_Hz, vibeLenArr_m, massPerLength_kgm, 
                                           stringDiameter_m, elasticity_Pa)

        # Calculate the deviation from the ideal frequencies in cents
        freqNewArr_Hz = freqArr_Hz * stiffFreqMultArr * tauFreqMultArr
        deltaFreq_cent = 1200 * np.log2(freqNewArr_Hz / freqIdealArr_Hz)
    
        return deltaFreq_cent
    
    return calc_deltafreq_cent
````

Test the model function by specifying offset nut and bridge and plotting the intonation error compared to a standard fretboard:

````python
# Geometric parameters of a ukulele
nFrets =               18
scaleLength_m =     367.0 / 1000
actionNut_m  =        1.5 / 1000
actionFret12_m =      3.5 / 1000
fingerWidth_m =       5.0 / 1000

# String specific parameters
freqOpen_Hz =       440.0
massPerLength_kgm =  1e-4
stringDiameter_m =    0.5 / 1000
elasticity_Pa =       4e9
nutOffset_m =        -2.0 / 1000
saddleOffset_m =      3.0 / 1000

# Get the model function
model = get_model_func(scaleLength_m, nFrets, actionNut_m, actionFret12_m, 
                                     fingerWidth_m, freqOpen_Hz, massPerLength_kgm, 
                                     stringDiameter_m, elasticity_Pa)

# Calculate for the offset nut & bridge
p = nutOffset_m, saddleOffset_m
deltaFreq_cent = model(p) 

# Calculate the uncorrected intonation
pU = [0, 0]
deltaFreqNoOff_cent = model(pU)

# Plot the frequency difference cents
ax = plt.subplot(1, 1, 1)
ax.plot(fretNumArr[1:], deltaFreq_cent[1:], marker="<", label="Corrected")
ax.plot(fretNumArr[1:], deltaFreqNoOff_cent[1:], marker="o", label="Uncorrected")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.axhline(color="grey")
leg = ax.legend()
ax.set_xlabel("Fret Number")
ax.set_ylabel("$\Delta f (cents)$")
````

Now we need to define a $\chi^2$ function

````python
# Return a function to calculate the current chi-squared
def get_chisq_func(scaleLength_m, nFrets, actionNut_m, actionFret12_m,
                   fingerWidth_m, freqOpen_Hz, massPerLength_kgm, 
                   stringDiameter_m, elasticity_Pa, weightArr=None):
    """
    Return a function to calculate chi-squared for a instrument &
    string combination given the nut and saddle offsets.
    """
    
    # Return function takes only a vector of free parameters
    def calc_chisq(p, fjac=None):
        """
        Function to caclulate chi-sq given nut & bridge offset.
        
        p = [nutOffset_m, saddleOffset_m]
        
        """
        status = 0
        
        # Assign vector to named variables
        #nutOffset_m, saddleOffset_m = p
        
        # Get the model function
        model = get_model_func(scaleLength_m, nFrets, actionNut_m, actionFret12_m, 
                               fingerWidth_m, freqOpen_Hz, massPerLength_kgm, 
                               stringDiameter_m, elasticity_Pa)
        
        # Calculate the intonation error array
        deltaFreq_cent = model(p)
        
        # Set the weight to unity by default
        if weightArr is None:
            weights = np.ones_like(deltaFreq_cent)
        else:
            weights = weightArr
        
        # Calculate chi-squared
        chiSq = np.nansum( (deltaFreq_cent[1:]**2/weights[1:])**2 )
    
        return status, chiSq
        
    return calc_chisq
````

````python
# Get the chisq function for the current parameters
chisq_func = get_chisq_func(scaleLength_m, nFrets, actionNut_m, actionFret12_m, 
                            fingerWidth_m, freqOpen_Hz, massPerLength_kgm, 
                            stringDiameter_m, elasticity_Pa)

# Calculate chi-squared for the offset nut & bridge
p = nutOffset_m, saddleOffset_m
status, chiSq = chisq_func(p) 

# Calculate the reduced chi-squared
degFree = nFrets - len(p) - 1
chiSqRed = chiSq / degFree

print("chi-Squared         = {:.1f}".format(chiSq))
print("chi-Squared Reduced = {:.1f}".format(chiSqRed))
print(status)
````
